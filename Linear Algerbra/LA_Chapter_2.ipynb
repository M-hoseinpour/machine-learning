### Sums and Scalar Multiples

We say that two matrices are <b>equal</b> if they have the same size (i.e., the same number of rows
and the same number of columns) and if their corresponding columns are equal, which
amounts to saying that their corresponding entries are equal.

If A and B are m * n matrices, then the <b>sum</b> A + B is the m * n matrix whose columns are the sums of
the corresponding columns in A and B.

If r is a scalar and A is a matrix, then the <b>scalar multiple</b> rA is the matrix whose columns are r times the corresponding columns in A.
