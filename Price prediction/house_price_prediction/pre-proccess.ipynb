```python
import pandas as pd
```

```python
HOUSING_PATH = "D:\house-prediction\house-price-prediction\dataset-tehran.csv"
housing = pd.read_csv(HOUSING_PATH)
```

```python
count_no_price = housing['price'].eq('توافقی').sum()
count_no_price
```

```python
housing = housing[housing['price'] != 'توافقی']
```

```python
housing = housing.dropna(subset=['id'])
```

```python
housing.dtypes
```

```python
numeric_columns = ['id', 'meterage', 'property_floor', 'area_meterage', 'number_of_units_per_floor', 'number_floor', 'year_of_building', 'lat', 'lon']
housing[numeric_columns] = housing[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
housing['price'] = pd.to_numeric(housing['price'].str.replace(',', ''), errors='coerce')
```

```python
categorical_columns = housing.select_dtypes(include='object').columns

categorical_columns
```

```python
housing.describe()
```

```python
housing.info()
```

```python
# features with less NAN values
main_features = [
    'age',
    'elevator',
    'parking',
    'rooms',
    "meterage",
    "year_of_building",
    "warehouse",
    'property_floor',
    "property_type",
    "lat",
    "lon"]
```

```python
housing['warehouse'].value_counts()
```

```python
housing['warehouse'] = housing['warehouse'].replace({'دارد': 1, 'ندارد': 0, 'true': 1})
housing["elevator"]  = housing["elevator"].replace({'دارد': 1, 'ندارد': 0, 'true': 1})
housing["parking"]   = housing["parking"].replace({'دارد': 1, 'ندارد': 0, 'true': 1})
housing['rooms']     = housing['rooms'].map({'بدون اتاق': 0, '1': 1, '2': 2, '3': 3, '4': 4, '5 به بالا': 5})
housing["property_type"] = housing["property_type"].replace({ "آپارتمان" : 1, "خانه و کلنگی" : 0 })
housing['age']   = housing['age'].map({ 'نوساز': 0, '۲-۵ سال': 1, '۵-۱۰ سال': 2, '۱۰-۱۵ سال': 3, '۱۵-۲۰ سال': 4, '۲۰ سال به بالا': 5})
```

```python
housing["warehouse"].value_counts()
```

```python
corr_matrix = housing[["price", "age", "elevator", "parking", "rooms", "meterage", "year_of_building", "warehouse", "property_floor", "property_type", "lat", "lon"]].corr()
corr_matrix["price"].sort_values(ascending=False)
```

```python
for column in main_features:
    mode_value = housing[column].mode().iloc[0]  # Calculate the mode and get the first value
    housing[column].fillna(mode_value, inplace=True)
```
